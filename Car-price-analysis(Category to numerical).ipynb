```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#Reading the file in pandas
car_sales = pd.read_csv("car-sales-extended.csv")
car_sales
```

```python
# assigning the values for training 
x = car_sales.drop('Price',axis = 1) # X is the feature (conatining all the datas except price)
y = car_sales["Price"] # Y is the target containing the price value 
x
```

```python
## Converting categorical to numerical  
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
category = ["Make",'Colour']
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,category)],remainder = "passthrough")
transformed_x = transformer.fit_transform(x)
transformed_x
```

```python
#Splitting for train and test validation : 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(transformed_x,y,test_size = 0.2)
```

```python
#Choosing the right machine learning model to train : 
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train) 
percentage = model.score(x_test,y_test)
print("It has a {} percentage of success".format(percentage*100))
```
